```python
import openai
from pydub import AudioSegment
import soundfile as sf
import librosa

# Set your OpenAI API key
openai.api_key = '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'

# Load your MP3 file
mp3_file_path = 'C:\\Users\\ESHOP\\Documents\\Share Mobility\\speech-to-text\\test_audio.mp3'
audio = AudioSegment.from_mp3(mp3_file_path)

# Convert MP3 to WAV format
wav_file_path = 'C:\\Users\\ESHOP\\Documents\\Share Mobility\\speech-to-text'
audio.export(wav_file_path, format="wav")

# Load the WAV file using librosa
audio_data, sample_rate = librosa.load(wav_file_path, sr=None)

# Ensure audio is in the required format (16kHz)
if sample_rate != 16000:
    audio_data = librosa.resample(audio_data, orig_sr=sample_rate, target_sr=16000)
    sample_rate = 16000

# Save the resampled audio
resampled_audio_file_path = 'resampled_audio.wav'
sf.write(resampled_audio_file_path, audio_data, sample_rate)

# Read the resampled audio file
with open(resampled_audio_file_path, "rb") as audio_file:
    audio_bytes = audio_file.read()

# Use the Whisper model to transcribe the audio
response = openai.Audio.transcribe(
    model="whisper-1",
    file=open(resampled_audio_file_path, "rb")
)

# Print the transcribed text
print(response['text'])
```
